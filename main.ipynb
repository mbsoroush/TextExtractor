# 1

```python
# Read from the file 'stsdev.txt'
with open('stsdev.txt', 'r', encoding='utf8') as f_r:
    read_file = f_r.readlines()

# Write to the file 'a.txt' and 'b.txt'
with open('a.txt', 'w', encoding='utf8') as f_a, open('b.txt', 'w', encoding='utf8') as f_b:
    for line in read_file:
        columns = line.split('\t')
        if len(columns) >= 2:  # Ensure there are at least two columns
            f_a.write(columns[-2] + '\n')
            f_b.write(columns[-1])

print("Files 'a.txt' and 'b.txt' have been written successfully.")
```

# 2

```python
# Open and read files a.txt and b.txt
with open('a.txt', 'r', encoding='utf-8') as file_a:
    read_a = file_a.read()
    
with open('b.txt', 'r', encoding='utf-8') as file_b:
    read_b = file_b.read()

# Extract words from both files
words_a = read_a.lower().strip().split()
words_b = read_b.lower().strip().split()

# Use a set to store unique words
unique_words = set(words_a + words_b)

# Write unique words to vocab.txt
with open('vocab.txt', 'w', encoding='utf-8') as unique:
    for word in sorted(unique_words):  # Sorting for a more readable output
        unique.write(word + '\n')
```

```python
vector_glove_numbers = open('vector_glove.pth', 'r', encoding="utf8").read().split(' ')
vocab_glove = open('vocab_glove.txt', 'r', encoding="utf8").readlines()

columns = 300
row = len(vocab_glove)
x = np.zeros((row,columns))
k = 0
for i in range(0,row):
    for j in range(0,columns):
        x[i][j] = float(vector_glove_numbers[k])
        k = k + 1
matrix_number = np.matrix(x)
np.savetxt("vector_glove.pth", matrix_number, fmt='%s')        
```

```python
import random

# Function to generate a list of 21 random numbers between -1 and 1
def generate_random_list():
    return [random.uniform(-1, 1) for _ in range(21)]

# Read vocab.txt and create the label2index dictionary
with open('vocab.txt', 'r', encoding='utf-8') as f_vocab:
    vocab_lines = f_vocab.readlines()

label2index = {word.strip(): [index, generate_random_list()] for index, word in enumerate(vocab_lines)}

# Read D2.txt and store its contents
with open('D2.txt', 'r', encoding='utf-8') as f_d2:
    d2_lines = f_d2.readlines()

# Create a dictionary from D2.txt with words as keys and their 21 numbers as values
d2_dict = {}
for line in d2_lines:
    parts = line.strip().split()
    word = parts[0]
    numbers = list(map(float, parts[1:22]))  # Assuming the first 21 numbers after the word
    d2_dict[word] = numbers

# Update label2index with numbers from D2.txt if the word exists in D2.txt
for word in label2index.keys():
    if word in d2_dict:
        label2index[word][1] = d2_dict[word]

# Output the updated label2index dictionary
print("label2index dictionary:", label2index)
```

```python
import random

# Function to generate a list of 21 random numbers between -1 and 1
def generate_random_list():
    return [random.uniform(-1, 1) for _ in range(21)]

# Read vocab.txt and create the label2index dictionary
with open('vocab.txt', 'r', encoding='utf-8') as f_vocab:
    vocab_lines = f_vocab.readlines()

label2index = {word.strip(): [index, generate_random_list()] for index, word in enumerate(vocab_lines)}

# Read D2.txt and store its contents
d2_dict = {}
with open('D2.txt', 'r', encoding='utf-8') as f_d2:
    for line in f_d2:
        parts = line.strip().split()
        word = parts[0]
        # Ensure that there are exactly 21 numbers after the word
        if len(parts) == 22:
            try:
                numbers = list(map(float, parts[1:22]))  # Convert the first 21 numbers after the word
                d2_dict[word] = numbers
            except ValueError:
                print(f"Skipping line due to conversion error: {line}")

# Update label2index with numbers from D2.txt if the word exists in D2.txt
for word in label2index.keys():
    if word in d2_dict:
        label2index[word][1] = d2_dict[word]

# Select a random line from b.txt and process it
with open('b.txt', 'r', encoding='utf-8') as f_b:
    b_lines = f_b.readlines()
    random_line_index = random.randint(0, len(b_lines) - 1)
    random_line = b_lines[random_line_index].strip()
    words_in_random_line = random_line.split()

# Create the list of 21 corresponding numbers for each word in the random line
numbers_list = []
for word in words_in_random_line:
    if word in label2index:
        numbers_list.extend(label2index[word][1])

# Save the content and the selected line number in b10.txt
with open('b10.txt', 'w', encoding='utf-8') as f_output:
    f_output.write(f"Selected line number: {random_line_index}\n")
    f_output.write(f"Selected line: {random_line}\n")
    f_output.write("Generated numbers: " + ' '.join(map(str, numbers_list)) + '\n')

# Output the generated list for verification
print(f"Selected line number: {random_line_index}")
print(f"Selected line: {random_line}")
print("Generated numbers:", numbers_list)
```
